## Model to diagnose Breast Cancer

### Based on a Support Vector Classifier trained on a dataset of cancerous and non cancerous cells with respective diagnosis.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Load data

df = pd.read_csv(r'C:\Users\91988\Downloads\archive (2)\data.csv')
```

```python
# Data types

df.info()
```

```python
# Check for null values

df.isnull().sum()
```

### Here B = Benign M = Malignant

```python
# Count Plot
df['diagnosis'] = df['diagnosis'].astype('category')
palette ={'B' : 'lightblue', 'M' : 'red'}
sns.countplot(data = df, x = 'diagnosis', palette = palette)
```

```python
# Corellation plot

df = df.drop(['Unnamed: 32', 'id'], axis = 1)
num_col = df.select_dtypes(include = ['float64', 'int64']).columns
corrdf = df[num_col].corr()
plt.figure(figsize = (20, 10))
sns.heatmap(corrdf, annot = True)
plt.show()
```

```python
# Splitting dataset

y = df['diagnosis'].values
x = df.drop('diagnosis', axis = 1).values
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Defining the SVC model

svc = SVC(probability = True)
svc.fit(x_train, y_train)
y_pred_svc = svc.predict(x_test)
```

```python
# Compare some predictions to actual values

print(y_test[0:10])
print(y_pred_svc[0:10])
```

### The defined model is very accurate as seen in its performance on the test set.

```python
# Checking accuracy of model

print(accuracy_score(y_pred_svc, y_test))
```

### Compare it to other model based on Kneighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Train model

#KNeighbors Classifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

```python
# Compare actual and predicted values

print(y_test[0:10])
print(y_pred_knn[0:10])
```

```python
# Accuracy Score

print(accuracy_score(y_pred_knn, y_test))
```

### It is slightly higher than that of SVC

## Classification Report

```python
# SVC
print(classification_report(y_test, y_pred_svc))
```

```python
# KNN
print(classification_report(y_test, y_pred_knn))
```
